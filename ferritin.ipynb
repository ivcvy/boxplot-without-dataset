```python
all_path = '/content/drive/MyDrive/ferritin/all.csv'
```

```python
import pandas as pd
all = pd.read_csv(all_path)
```

```python
all
```

```python
all[["med"]] = all[["med"]].astype(float)
all[["q1"]] = all[["q1"]].astype(float)
all[["q3"]] = all[["q3"]].astype(float)
all[["whislo"]] = all[["whislo"]].astype(float)
all[["whishi"]] = all[["whishi"]].astype(float)
```

```python
import numpy as np

all[["med"]] = np.log10(all[["med"]])
all[["q1"]] = np.log10(all[["q1"]])
all[["q3"]] = np.log10(all[["q3"]])
all[["whislo"]] = np.log10(all[["whislo"]])
all[["whishi"]] = np.log10(all[["whishi"]])
```

```python
all1 = all[["med","q1","q3","whislo","whishi"]]
dict1 = all1.to_dict(orient='index')
allfinal = dict1.values()
allfinal
```

```python
ylabel = all[["label"]]
ser1 = ylabel.squeeze()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ylabel_list = ser1.tolist()
xlabel_list = ["1", "10", "100", "1,000", "10,000", "100,000"]

fig, ax = plt.subplots();
ax.bxp(allfinal, showfliers=False, vert=False);

x = ax.get_xticks()
x_tick = np.arange(0, max(x), 1)
plt.xticks(ticks = x_tick, labels = xlabel_list)
ax.set_xlabel("Serum ferritin (ng/mL)")

y = ax.get_yticks()
plt.yticks(ticks = y, labels = ylabel_list)

plt.rcParams["figure.figsize"] = [1, 1]
plt.rcParams["figure.autolayout"] = True
```

```python
fig.savefig("all1.png")
```
